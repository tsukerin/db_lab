<b>Импорт библиотек</b>

```python
import pandas as pd
import re
pd.options.mode.chained_assignment = None

df = pd.read_excel('Машина 1.xlsx')
df
```

<b>Задание 1</b>

```python
df['Н'] = df['Наименование компоненты']
assembly = ['Кузов', 'Салон', 'Подвеска', 'Двигатель', 'Трансмиссия']

for column, series in df.items():
    if column == 'Н':
        for i in range(len(series)):
            if series[i] in assembly: df.at[i, column] = 'Сборка'
            else: df.at[i, column] = 'Компонент'
            
df
```

<b>Задание 2</b>

```python
fastener = [ 'Болт', 'Гайка', 'Штифт', 'Шайба', 'Шуруп' ]
decor = [ 'Коврики', 'Подушки' ]

df['I'] = df['Наименование компоненты']

for column, series in df.items():
    if column == 'I':
        for i in range(len(series)):
            if series[i] in fastener: series[i] = 'Крепеж'
            elif series[i] in decor: series[i] = 'Декор'
            elif series[i] in assembly: series[i] = 'Сборка'
            else: series[i] = 'Детали'

df
```

<b>Задание 3</b>

```python
df['J'] = df['Код компоненты']

for column, series in df.items():
    if column == 'J':
        for i in range(len(series)):
            if bool(re.search('[a-zA-Z]', series[i])) == True: series[i] = 'Иностранное'
            elif bool(re.search('[а-яА-Я]', series[i])) == True and not series[i].__contains__('ГОСТ'): series[i] = 'РФ'
            elif series[i].__contains__('ГОСТ'): series[i] = 'ГОСТ'

df
```

<b>Экспорт excel файла 'Машина 1-разметка.xlsx'</b>

```python
df.to_excel('Машина 1-разметка.xlsx', index=False)
```
