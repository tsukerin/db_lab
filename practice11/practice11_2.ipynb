```python
import pandas as pd

df1 = pd.read_excel('Машина 1.xlsx')
df2 = pd.read_excel('Машина 2.xlsx')

changes_df = pd.DataFrame(columns=df2.columns.tolist() + ['Изменения'])

for index, row in df2.iterrows():
    matching_row = df1[(df1['Код сборки'] == row['Код сборки']) &
                       (df1['Наименование сборки'] == row['Наименование сборки']) &
                       (df1['Код компоненты'] == row['Код компоненты']) &
                       (df1['Наименование компоненты'] == row['Наименование компоненты'])]

    if not matching_row.empty:
        if matching_row['Кол-во на сборку'].values[0] != row['Кол-во на сборку']:
            changes_df = pd.concat([changes_df, row.to_frame().T.assign(Изменения='Изменено количество')], ignore_index=True)
        else:
            changes_df = pd.concat([changes_df, row.to_frame().T.assign(Изменения='Изменений не было')], ignore_index=True)
    else:
        changes_df = pd.concat([changes_df, row.to_frame().T.assign(Изменения='Элемент добавлен')], ignore_index=True)

for index, row in df1.iterrows():
    matching_row = df2[(df2['Код сборки'] == row['Код сборки']) &
                       (df2['Наименование сборки'] == row['Наименование сборки']) &
                       (df2['Код компоненты'] == row['Код компоненты']) &
                       (df2['Наименование компоненты'] == row['Наименование компоненты'])]

    if matching_row.empty:
        changes_df = pd.concat([changes_df, row.to_frame().T.assign(Изменения='Элемент удален')], ignore_index=True)

changes_df
```

```python
changes_df.to_excel('Машина 2-изменения.xlsx', index=False)
```
